```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculer la popularitÃ© totale par genre principal
main_genre_popularity = df.groupby('main_genre')['popularity'].mean().sort_values(ascending=False)

# CrÃ©er un bar plot pour les genres principaux
plt.figure(figsize=(12, 6))
sns.barplot(x=main_genre_popularity.index, y=main_genre_popularity.values, palette='viridis')
plt.title('Popularity per genre', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
